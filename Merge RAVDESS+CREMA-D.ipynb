```python


# Concatenate training data
X_train = np.concatenate((X_train_R, X_train_CD1, X_train_CD2, X_train_CD3), axis=0)
y_train = np.concatenate((y_train_R, y_train_CD1, y_train_CD2, y_train_CD3), axis=0)

# Concatenate validation data
X_valid = np.concatenate((X_valid_R, X_valid_CD1, X_valid_CD2, X_valid_CD3), axis=0)
y_valid = np.concatenate((y_valid_R, y_valid_CD1, y_valid_CD2, y_valid_CD3), axis=0)

# Concatenate testing data
X_test = np.concatenate((X_test_R, X_test_CD1, X_test_CD2, X_test_CD3), axis=0)
y_test = np.concatenate((y_test_R, y_test_CD1, y_test_CD2, y_test_CD3), axis=0)

# Check merged shapes
print(f'Merged X_train: {X_train.shape}, Merged y_train: {y_train.shape}')
print(f'Merged X_valid: {X_valid.shape}, Merged y_valid: {y_valid.shape}')
print(f'Merged X_test: {X_test.shape}, Merged y_test: {y_test.shape}')
```

```python
###### SAVE #########
# choose save file name
filename = 'CREMA-D + RAVDESS.npy'
path = '/content/gdrive/My Drive/thesis/Thesis/Pre Data/'
fullPath = path+filename
# open file in write mode and write data
with open(fullPath, 'wb') as f:
    np.save(f, X_train)
    np.save(f, X_valid)
    np.save(f, X_test)
    np.save(f, y_train)
    np.save(f, y_valid)
    np.save(f, y_test)

print(f'Features and labels saved to {fullPath}')
```
