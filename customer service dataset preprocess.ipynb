```python
text_data = ["Hello there! It's a wonderful day!", "I'm so happy to see you."]
emotion = "happy"
```

```python
from TTS.api import TTS
```

```python


# Load a pre-trained TTS model with emotion control
# You can explore models at https://github.com/coqui-ai/TTS#available-models
tts = TTS(model_name="tts_models/multilingual/multi-dataset/your_tts")
```

```python
# Print the list of available speakers
print("Available speakers:", tts.speakers)
```

```python
for idx, text in enumerate(text_data):
    print(f"Synthesizing audio for: {text} with emotion: {emotion}")
    # Generate audio with emotion
    audio_path = f"output_{emotion.lower()}_{idx + 1}.wav"
    tts.tts_to_file(
        text=text,
        file_path=audio_path,
        speaker_wav='female-en-5',  # Optional: Provide a reference audio for speaker cloning
        #style_wav=None,  # Optional: Provide a style reference for emotion
        style_emotion=emotion,  # Use emotion tag if supported by the model
        language="en",

    )
    print(f"Audio saved to {audio_path}")
```
