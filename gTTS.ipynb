```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/thesis/Thesis/Datasets/customer_support_tickets.csv')
df_subset = df[['Ticket Description', 'Resolution']]
df_subset.head()
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(df_subset)

# Define a list of possible emotions
emotions = ["happy", "sad", "angry", "frustrated", "neutral", "excited"]

# Randomly assign an emotion to each row
np.random.seed(42)  # for reproducible results, optional
df_subset["Emotion"] = np.random.choice(emotions, size=len(df))

df_subset
```

```python
df_subset['Resolution'] = df_subset['Resolution'].replace('NaN', pd.NA)  # Replace 'Null' with pandas' NA value
df = df_subset.dropna(subset=['Resolution']).reset_index(drop=True)

df
```

```python
import torch
from parler_tts import ParlerTTSForConditionalGeneration
from transformers import AutoTokenizer
import soundfile as sf

device = "cuda:0" if torch.cuda.is_available() else "cpu"

model = ParlerTTSForConditionalGeneration.from_pretrained("parler-tts/parler-tts-mini-v1").to(device)
tokenizer = AutoTokenizer.from_pretrained("parler-tts/parler-tts-mini-v1")


```

```python

# Define the prompt and description
prompt = "Today was a bad day. I broke off with my boy friend. Give me some company."
description = "A female speaker delivers the speech in a calm and neutral tone, with a balanced and even pace, maintaining a professional and unemotional demeanor. The audio is very clear, with no background noise, ensuring a straightforward and clear delivery."

# Tokenize the inputs and move them to the appropriate device
inputs = tokenizer(description, return_tensors="pt").to(device)
prompt_inputs = tokenizer(prompt, return_tensors="pt").to(device)

# Generate the audio
  # Disable gradient calculation for inference
generation = model.generate(
        input_ids=inputs.input_ids,
         # Attention mask for the description
        prompt_input_ids=prompt_inputs.input_ids
         # Attention mask for the prompt
    )

# Convert the generated audio to a numpy array and save it as a WAV file
audio_arr = generation.cpu().numpy().squeeze()
sf.write("parler_tts_out.wav", audio_arr, model.config.sampling_rate)

print("Audio generated successfully and saved as 'parler_tts_out.wav'.")
```

```python
for index, row in df.iterrows():
    prompt = row[0]
    emotion = row[2]

    if emotion == "happy":
        description = "A female speaker delivers the speech in a cheerful and upbeat tone, with a lively and energetic pace, radiating positivity and joy. The audio is very clear, with a bright and warm quality, enhancing the sense of happiness."
    elif emotion == "sad":
        description = "A female speaker delivers the speech in a somber and melancholic tone, with a slow and deliberate pace, conveying deep emotion and sorrow. The audio is very clear, with minimal background noise, emphasizing the emotional weight of the delivery."
    elif emotion == "angry":
        description_angry = "A female speaker delivers the speech in an intense and aggressive tone, with a fast and forceful pace, clearly expressing anger and frustration. The audio is very clear, with a sharp and loud quality, emphasizing the intensity of the emotion."
    elif emotion == "frustrated":
        description = "A female speaker delivers the speech in an irritated and impatient tone, with a slightly faster pace and occasional pauses to convey frustration. The audio is very clear, with a sharp and intense quality, reflecting the speaker's annoyance."
    elif emotion == "neutral":
        description = "A female speaker delivers the speech in a soothing and relaxed tone, with a slow and steady pace, conveying a sense of peace and tranquility. The audio is very clear, with a soft and warm quality, creating a calming atmosphere."
    elif emotion == "excited":
        description = "A female speaker delivers the speech in an energetic and enthusiastic tone, with a fast and lively pace, radiating excitement and joy. The voice is bright, expressive, and slightly higher in pitch, conveying a sense of anticipation and eagerness. The audio is very clear, with minimal background noise, enhancing the vibrant and dynamic quality of the delivery."

    inputs = tokenizer(description, return_tensors="pt").to(device)
    prompt_inputs = tokenizer(prompt, return_tensors="pt").to(device)

    # Generate the audio
    # Disable gradient calculation for inference
    generation = model.generate(
            input_ids=inputs.input_ids,
            # Attention mask for the description
            prompt_input_ids=prompt_inputs.input_ids
            # Attention mask for the prompt
        )

    # Convert the generated audio to a numpy array and save it as a WAV file
    audio_arr = generation.cpu().numpy().squeeze()
    filename = f"audio data/{index}-{emotion}.wav"
    sf.write(filename, audio_arr, model.config.sampling_rate)

    print("Audio generated successfully and saved as ", filename)
```
